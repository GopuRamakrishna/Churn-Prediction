Checking for multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
X=add_constant(df.drop('Churn_Yes',axis=1))
vif=pd.DataFrame()
vif['Feature']=X.columns
vif['VIF']=[variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
vif
```

Train_test_split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop('Churn',axis=1),df['Churn'],test_size=0.3,random_state=42)
x_train.head()
```

```python
y_train[:5]
```

```python
df.columns
```

Normalise the data

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
pd.Series(y_train).value_counts()
```

as the data is imbalance using SMOTE for balancing the data

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=42)
x_train_resampled,y_train_resampled=smote.fit_resample(x_train_scaled,y_train)
```

```python
pd.Series(y_train_resampled).value_counts()
```

```python
x_train_resampled
```

applying the data for LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
model_log=LogisticRegression()
param_grid_logreg = {
    'penalty': ['l1', 'l2'],
    'C': [0.01, 0.1, 1, 10, 100],  # Regularization strength (inverse)
    'solver': ['liblinear'], 
    'max_iter': [100, 500, 1000],
}
grid=GridSearchCV(model_log,param_grid=param_grid_logreg,scoring='accuracy',cv=10,n_jobs=-1,verbose=2)
```

```python
grid.fit(x_train_resampled,y_train_resampled)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
best_log_model=LogisticRegression(C= 100, max_iter= 100, penalty= 'l1', solver= 'liblinear')
best_log_model.fit(x_train_resampled,y_train_resampled)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,best_log_model.predict(x_test_scaled))
```

Applying the model for RandomForestClassifier


```python
from sklearn.ensemble import RandomForestClassifier
model_rf=RandomForestClassifier()
param_grid={
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'bootstrap': [True, False]
}
grid_rf = GridSearchCV(
    model_rf,
    param_grid=param_grid,
    scoring='accuracy',
    cv=10,
    n_jobs=-1,
    verbose=2
)
```

```python
grid_rf.fit(x_train_resampled,y_train_resampled)
```

```python
grid_rf.best_params_
```

```python
grid_rf.best_score_
```

```python
best_rf_model = RandomForestClassifier(
    bootstrap=True,
    max_depth=None,
    min_samples_leaf=1,
    min_samples_split=2,
    n_estimators=100,
    random_state=42  # Optional: for reproducibility
)
best_rf_model.fit(x_train_resampled,y_train_resampled)
accuracy_score(y_test,best_rf_model.predict(x_test_scaled))
```

applying the dataset for the Xgboost

```python
from xgboost import XGBClassifier
xgb_model = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')

param_grid_xgb = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0],
    'gamma': [0, 1, 5]
}

grid_xgb = GridSearchCV(
    xgb_model,
    param_grid=param_grid_xgb,
    scoring='accuracy',  
    cv=10,
    n_jobs=-1,
    verbose=2
)

```

```python
grid_xgb.fit(x_train_resampled, y_train_resampled)

print("Best XGB Params:", grid_xgb.best_params_)
print("Best XGB Accuracy:", grid_xgb.best_score_)
```

```python
best_xgb_model = XGBClassifier(
    colsample_bytree=0.8,
    gamma=0,
    learning_rate=0.01,
    max_depth=5,
    n_estimators=100,
    subsample=0.8,
    use_label_encoder=False,  # Disable label encoder (for newer versions)
    eval_metric='logloss',    # Recommended to suppress warning
    random_state=42           # Optional: for reproducibility
)
```

```python
best_xgb_model.fit(x_train_resampled,y_train_resampled)
accuracy_score(y_test,best_xgb_model.predict(x_test_scaled))
```

```python
best_model=best_xgb_model
```

dump the model
